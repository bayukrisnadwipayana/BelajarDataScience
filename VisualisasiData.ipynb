```python
#baca dataset
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("climate_change.csv")
df.rename(columns={'date':'tanggal','meantemp':'suhu','humidity':'kelembapan','wind_speed':'kecepatan angin','meanpressure':'tekanan udara'},inplace=True)
df
```

```python
#cek tipe data setiap kolom
df.dtypes
```

```python
#ubah tipe data tanggal menjadi datetime
df.astype({'tanggal':'datetime64[ns]'})
```

```python
df.dtypes
```

```python
#deskripsi statistik dataset
df.describe()
```

```python
#ambil data kelembapan 10 data terakhir dan tampilkan
df.kelembapan.tail(10)

#cek data missing value
df.isnull().sum()

#ambil 3 data
df.head(12)

#bining data suhu pengelompokan data berdasarkan label
label=['sejuk','panas']
df['bining']=pd.qcut(df['suhu'],q=2,labels=label)
df[df['bining']=='panas']
```

```python
#visualisasikan hubungan antara suhu dan kelembapan
fig=plt.figure(figsize=(12,10))
x_suhu=df['suhu']
y_kelembapan=df['kelembapan']
y_angin=df['kecepatan angin']
y_tekanan=df['tekanan udara']
plt.plot(x_suhu,linestyle='dotted',color='blue')
plt.plot(y_kelembapan,linestyle='dotted',color='red')
plt.plot(y_angin,linestyle='dotted',color='yellow')
plt.show()
```
